```python
#Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import warnings

warnings.filterwarnings("ignore")
```

```python
# Load the dataset
df = pd.read_csv('customer_data.csv')
```

```python
df.head()
```

```python
#set the styles for plots
sns.set(style='whitegrid')
```

```python
# 1.Histogram with Kde

plt.figure(figsize=(6, 4))
# Plot histogram with KDE for Age
sns.histplot(data=df, x='Age', kde=True, color='skyblue', bins=5)

# Add labels and title
plt.title('Age Distribution of Customers')
plt.xlabel('Age')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

```python
# 2. Scatter plot with hue

# Add 'Gender' column to the data
data = {
    'CustomerID': [1, 2, 3, 4, 5],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, 30, 28, 35, 22],
    'Email': [
        'alice@example.com',
        'bob@example.com',
        'charlie@example.com',
        'david@example.com',
        'eva@example.com'
    ],
    'Gender': ['Female', 'Male', 'Male', 'Male', 'Female']  # <-- added
}

df = pd.DataFrame(data)

# Now plot
sns.scatterplot(data=df, x='CustomerID', y='Age', hue='Gender', s=100)

# Labels
plt.title('Customer Age by Gender')
plt.xlabel('Customer ID')
plt.ylabel('Age')
plt.show()
```

```python

df = pd.DataFrame(data)

# Create box plot
sns.boxplot(data=df, x='Gender', y='Age')

# Add title and labels
plt.title('Age Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.show()
```

```python
# Sample data
data = {
    'CustomerID': [1, 2, 3, 4, 5],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, 30, 28, 35, 22],
    'Email': [
        'alice@example.com',
        'bob@example.com',
        'charlie@example.com',
        'david@example.com',
        'eva@example.com'
    ]
}

df = pd.DataFrame(data)

# Create bar plot
sns.barplot(data=df, x='Name', y='Age', palette='Set2')

# Add title and labels
plt.title('Customer Age by Name')
plt.xlabel('Customer Name')
plt.ylabel('Age')
plt.show()
```

```python
# pair plot
# Sample data
data = {
    'CustomerID': [1, 2, 3, 4, 5],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, 30, 28, 35, 22],
    'Email': [
        'alice@example.com',
        'bob@example.com',
        'charlie@example.com',
        'david@example.com',
        'eva@example.com'
    ]
}

df = pd.DataFrame(data)

# Since 'Name' and 'Email' are non-numeric, they are not useful in a pair plot
# Creating a pair plot with only numeric columns
sns.pairplot(df[['CustomerID', 'Age']])

plt.suptitle('Pair Plot of Customer Data', y=1.02)
plt.show()
```

```python
# 6. heatmap
# Sample data
data = {
    'CustomerID': [1, 2, 3, 4, 5],
    'Age': [25, 30, 28, 35, 22],
}

df = pd.DataFrame(data)

# Create a correlation matrix
corr_matrix = df.corr()

# Plot the heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()

```
